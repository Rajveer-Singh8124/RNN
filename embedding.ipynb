```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sentences=[  

 "The sun rose early today.",  
 "She smiled at the puppy."
, "The rain fell softly."  
,"He opened the old book."  
,"Birds sang in the trees."  
, "They walked by the lake."  
,"A breeze cooled the air."  
, "The stars lit up the night."  ]
```

```python
## define the vocabulary
voc_size = 10000
```

```python
###One hot representation
one_hot_rep = [one_hot(words,voc_size) for words in sentences]
one_hot_rep
```

```python
# word embedding
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
sentences_len = 7
embedded_docs = pad_sequences(one_hot_rep,maxlen=sentences_len)
embedded_docs
```

```python
## feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length=sentences_len))
model.compile("adam","mse")
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
